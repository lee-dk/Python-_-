```python
import pandas as pd
import seaborn as sns
import numpy as np
import random

iris = sns.load_dataset("iris")
iris_x = iris.loc[:, ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]

random.seed(1)
for col in range(len(iris_x.columns)):
    iris_x.iloc[[random.sample(range(len(iris)), 20)], col] = float('NaN')
display('데이터프레임 원본', iris_x.head(10))

# (1) 누락데이터 행 제거
df = iris_x.head(10).copy()
df.isnull().sum(axis=0) # 각 열에서 누락데이터 개수 확인
df_result = df.dropna(axis=0)
display('(1) 누락데이터 행 제거', df_result)

# (2) 행 제거, 추가
df.drop(df.index[7], inplace=True)
df.loc[10] = [5.4, 3.7, 1.5, 0.2]
display('(2) 행 제거, 추가', df)

# (3) 행 추가 후 sepal_length, sepal_width 열 기준 결측치(NaN)있는 행 제거
df.loc[11] = [4.8, 3.4, 1.6, np.nan]
df.loc[12] = [4.8, 3.0, 1.4, 0.1]
df.loc[13] = [4.3, 3.0, 1.1, 0.1]
df.loc[14] = [5.8, 4.0, 1.2, 0.2]

df.dropna(subset=['sepal_length', 'sepal_width'], how='any', axis=0, inplace=True)
display('(3) 행 추가 후 sepal_length, sepal_width 열 기준 결측치(NaN)있는 행 제거', df)

# (4) 결측치 0으로 치환
df = iris_x.head(10).copy()
df.fillna(0, inplace=True)
display('(4) 결측치 0으로 치환', df)

# (5) 결측치 특정 값으로 치환
df = iris_x.head(10).copy()
df['sepal_length'].fillna(5.8, inplace=True)
df['sepal_width'].fillna(3.1, inplace=True)
df['petal_length'].fillna(3.7, inplace=True)
df['petal_width'].fillna(1.2, inplace=True)
display('(5) 결측치 특정 값으로 치환', df)

# (6) 결측치 10.0으로 치환
df = iris_x.head(10).copy()
df.fillna(10, inplace=True)
display('(6) 결측치 10으로 치환', df)

# (7) 결측치 이전 행 값으로 변경
df = iris_x.head(10).copy()
df.fillna(method='ffill', inplace=True)
display('(7) 결측치 이전 행 값으로 변경', df)
```

```python
import pandas as pd
import datetime

# (1) 일 간격으로 Timestamp 배열 생성
oct_day = pd.date_range(start='2020-10-01', end=None, periods=31, freq='D')
print('(1) 일 간격으로 Timestamp 배열 생성', oct_day, sep="\n")

# (2) 월 간격으로 Timestamp 배열 생성
year_2020 = pd.date_range(start='2020-01-31', end=None, periods=12, freq='M')
print('(2) 월 간격으로 Timestamp 배열 생성', year_2020, sep="\n")

# (3) 일 간격으로 Timestamp 배열 생성
oct_nov = pd.date_range(start='2020-10-01', end=None, periods=50, freq='D')
print('(3) 일 간격으로 Timestamp 배열 생성', oct_nov, sep="\n")

# (4) 1시간 간격으로 Timestamp 배열 생성
now = datetime.datetime.now()
today = now.strftime('%Y-%m-%d') # str

hour = pd.date_range(start=today, end=None, periods=24, freq='H')
print('(4) 1시간 간격으로 Timestamp 배열 생성', hour, sep="\n")

# (5) 시간 간격으로 Timestamp 배열 생성
hour_3 = pd.date_range(start=today, end=None, periods=8, freq='3H')
print('(5) 3시간 간격으로 Timestamp 배열 생성', hour_3, sep="\n")

# (6) 2달 간격으로 Timestamp 배열 생성
month_2020_2021 = pd.date_range(start='2020-01-31', end=None, periods=8, freq='2M')
print('(6) 2달 간격으로 Timestamp 배열 생성', month_2020_2021, sep="\n")
```

```python
import pandas as pd

# (1) log 파일 읽어오기
df = pd.read_table('./data/product_click.log', sep=" ")
display(df.head())
print(df.info())
print('\n')

# (2) 헤더 행 추가
df.columns = ['oldclicktime', 'clickpid']
display(df.head())
print('\n')

# (3) 자료형 변환
df['oldclicktime'] = df['oldclicktime'].astype('str')
print(df.info())
print('\n')

# (4) 문자열을 Timestamp로 변환
df['newclicktime'] = pd.to_datetime(df['oldclicktime'])
print(df.info())
display(df.head())
print('\n')

# (5) 날짜 데이터 분리
df['year'] = df['newclicktime'].dt.year
df['month'] = df['newclicktime'].dt.month
df['day'] = df['newclicktime'].dt.day
df['minute'] = df['newclicktime'].dt.minute
print(df.info())
display(df.head())
print('\n')
```
