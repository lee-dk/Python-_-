```python
# ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ìž„í¬íŠ¸í•©ë‹ˆë‹¤.
import cv2
import numpy as np
from matplotlib import pyplot as plt

# ì´ë¯¸ì§€ë¥¼ ë¡œë“œí•©ë‹ˆë‹¤.
image_bgr = cv2.imread('images/plane_256x256.jpg')

# BGRì—ì„œ HSVë¡œ ë³€í™˜í•©ë‹ˆë‹¤.
image_hsv = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2HSV)

# HSVì—ì„œ íŒŒëž‘ ê°’ì˜ ë²”ìœ„ë¥¼ ì •ì˜í•©ë‹ˆë‹¤.
lower_blue = np.array([50,100,50])
upper_blue = np.array([130,255,255])

# ë§ˆìŠ¤í¬ë¥¼ ë§Œë“­ë‹ˆë‹¤.
mask = cv2.inRange(image_hsv, lower_blue, upper_blue)

# ì´ë¯¸ì§€ì— ë§ˆìŠ¤í¬ë¥¼ ì ìš©í•©ë‹ˆë‹¤.
image_bgr_masked = cv2.bitwise_and(image_bgr, image_bgr, mask=mask)

# BGRì—ì„œ RGBë¡œ ë³€í™˜í•©ë‹ˆë‹¤.
image_rgb = cv2.cvtColor(image_bgr_masked, cv2.COLOR_BGR2RGB)

# ì´ë¯¸ì§€ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤.
plt.imshow(image_rgb), plt.axis("off")
plt.show()
```

```python
# ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ìž„í¬íŠ¸í•©ë‹ˆë‹¤.
import cv2
import numpy as np
from matplotlib import pyplot as plt

# ì´ë¯¸ì§€ë¥¼ ë¡œë“œí•˜ê³  RGBë¡œ ë³€í™˜í•©ë‹ˆë‹¤.
image_bgr = cv2.imread('images/plane_256x256.jpg')
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# ì‚¬ê°í˜• ì¢Œí‘œ: ì‹œìž‘ì ì˜ x, ì‹œìž‘ì ì˜ y, ë„ˆë¹„, ë†’ì´
rectangle = (0, 56, 256, 150)

# ì´ˆê¸° ë§ˆìŠ¤í¬ë¥¼ ë§Œë“­ë‹ˆë‹¤.
mask = np.zeros(image_rgb.shape[:2], np.uint8)

# grabCutì— ì‚¬ìš©í•  ìž„ì‹œ ë°°ì—´ì„ ë§Œë“­ë‹ˆë‹¤.
bgdModel = np.zeros((1, 65), np.float64)
fgdModel = np.zeros((1, 65), np.float64)

# grabCutì„ ì‹¤í–‰í•©ë‹ˆë‹¤.
cv2.grabCut(image_rgb, # ì›ë³¸ ì´ë¯¸ì§€
            mask, # ë§ˆìŠ¤í¬
            rectangle, # ì‚¬ê°í˜•
            bgdModel, # ë°°ê²½ì„ ìœ„í•œ ìž„ì‹œ ë°°ì—´
            fgdModel, # ì „ê²½ì„ ìœ„í•œ ìž„ì‹œ ë°°ì—´
            5, # ë°˜ë³µ íšŸìˆ˜
            cv2.GC_INIT_WITH_RECT) # ì‚¬ê°í˜•ì„ ì‚¬ìš©í•œ ì´ˆê¸°í™”

# ë°°ê²½ì¸ ê³³ì€ 0, ê·¸ì™¸ì—ëŠ” 1ë¡œ ì„¤ì •í•œ ë§ˆìŠ¤í¬ë¥¼ ë§Œë“­ë‹ˆë‹¤.
mask_2 = np.where((mask==2) | (mask==0), 0, 1).astype('uint8')

# ì´ë¯¸ì§€ì— ìƒˆë¡œìš´ ë§ˆìŠ¤í¬ë¥¼ ê³±í•´ ë°°ê²½ì„ ì œì™¸í•©ë‹ˆë‹¤.
image_rgb_nobg = image_rgb * mask_2[:, :, np.newaxis]

# ì´ë¯¸ì§€ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤.
plt.imshow(image_rgb_nobg), plt.axis("off")
plt.show()
```

```python
# ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ìž„í¬íŠ¸í•©ë‹ˆë‹¤.
import cv2
import numpy as np
from matplotlib import pyplot as plt

# í‘ë°± ì´ë¯¸ì§€ë¡œ ë¡œë“œí•©ë‹ˆë‹¤.
image_gray = cv2.imread("images/plane_256x256.jpg", cv2.IMREAD_GRAYSCALE)

# í”½ì…€ ê°•ë„ì˜ ì¤‘ê°„ê°’ì„ ê³„ì‚°í•©ë‹ˆë‹¤.
median_intensity = np.median(image_gray)

# ì¤‘ê°„ í”½ì…€ ê°•ë„ì—ì„œ ìœ„ì•„ëž˜ 1 í‘œì¤€ íŽ¸ì°¨ ë–¨ì–´ì§„ ê°’ì„ ìž„ê³„ê°’ìœ¼ë¡œ ì§€ì •í•©ë‹ˆë‹¤.
lower_threshold = int(max(0, (1.0 - 0.33) * median_intensity))
upper_threshold = int(min(255, (1.0 + 0.33) * median_intensity))

# ìºë‹ˆ ê²½ê³„ì„  ê°ì§€ê¸°ë¥¼ ì ìš©í•©ë‹ˆë‹¤.
image_canny = cv2.Canny(image_gray, lower_threshold, upper_threshold)

# ì´ë¯¸ì§€ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤.
plt.imshow(image_canny, cmap="gray"), plt.axis("off")
plt.show()
```

```python
# ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ìž„í¬íŠ¸í•©ë‹ˆë‹¤.
import cv2
import numpy as np
from matplotlib import pyplot as plt

# í‘ë°± ì´ë¯¸ì§€ë¡œ ë¡œë“œí•©ë‹ˆë‹¤.
image_bgr = cv2.imread("images/plane_256x256.jpg")
image_gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
image_gray = np.float32(image_gray)

# ëª¨ì„œë¦¬ ê°ì§€ ë§¤ê°œë³€ìˆ˜ë¥¼ ì„¤ì •í•©ë‹ˆë‹¤.
block_size = 2
aperture = 29
free_parameter = 0.04

# ëª¨ì„œë¦¬ë¥¼ ê°ì§€í•©ë‹ˆë‹¤.
detector_responses = cv2.cornerHarris(image_gray,
                                      block_size,
                                      aperture,
                                      free_parameter)

# ëª¨ì„œë¦¬ í‘œì‹œë¥¼ ë¶€ê°ì‹œí‚µë‹ˆë‹¤.
detector_responses = cv2.dilate(detector_responses, None)

# ìž„ê³„ê°’ë³´ë‹¤ í° ê°ì§€ ê²°ê³¼ë§Œ ë‚¨ê¸°ê³  í°ìƒ‰ìœ¼ë¡œ í‘œì‹œí•©ë‹ˆë‹¤.
threshold = 0.02
image_bgr[detector_responses >
          threshold *
          detector_responses.max()] = [255,255,255]

# í‘ë°±ìœ¼ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.
image_gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

# ì´ë¯¸ì§€ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤.
plt.imshow(image_gray, cmap="gray"), plt.axis("off")
plt.show()
```

```python
# ê°€ëŠ¥ì„±ì´ ë†’ì€ ëª¨ì„œë¦¬ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤.
plt.imshow(detector_responses, cmap='gray'), plt.axis("off")
plt.show()
```

```python
# ì´ë¯¸ì§€ë¥¼ ë¡œë“œí•©ë‹ˆë‹¤.
image_bgr = cv2.imread('images/plane_256x256.jpg')
image_gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

# ê°ì§€í•  ëª¨ì„œë¦¬ ê°œìˆ˜
corners_to_detect = 10
minimum_quality_score = 0.05
minimum_distance = 25

# ëª¨ì„œë¦¬ë¥¼ ê°ì§€í•©ë‹ˆë‹¤.
corners = cv2.goodFeaturesToTrack(image_gray,
                                  corners_to_detect,
                                  minimum_quality_score,
                                  minimum_distance)
corners = np.float32(corners)

# ëª¨ì„œë¦¬ë§ˆë‹¤ í° ì›ì„ ê·¸ë¦½ë‹ˆë‹¤.
for corner in corners:
    x, y = corner[0]
    cv2.circle(image_bgr, (x,y), 10, (255,255,255), -1)

# í‘ë°± ì´ë¯¸ì§€ë¡œ ë³€í™˜í•©ë‹ˆë‹¤.
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

# ì´ë¯¸ì§€ë¥¼ ì¶œë ¥í•©ë‹ˆë‹¤.
plt.imshow(image_rgb, cmap='gray'), plt.axis("off")
plt.show()
```

```python
# ë¼ì´ë¸ŒëŸ¬ë¦¬ë¥¼ ìž„í¬íŠ¸í•©ë‹ˆë‹¤.
import cv2
import numpy as np
from matplotlib import pyplot as plt

# ì´ë¯¸ì§€ë¥¼ ë¡œë“œí•©ë‹ˆë‹¤.
image_bgr = cv2.imread("images/plane_256x256.jpg", cv2.IMREAD_COLOR)

# RGBë¡œ ë³€í™˜í•©ë‹ˆë‹¤.
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

# íŠ¹ì„± ê°’ì„ ë‹´ì„ ë¦¬ìŠ¤íŠ¸ë¥¼ ë§Œë“­ë‹ˆë‹¤.
features = []

# ê° ì»¬ëŸ¬ ì±„ë„ì— ëŒ€í•´ ížˆìŠ¤í† ê·¸ëž¨ì„ ê³„ì‚°í•©ë‹ˆë‹¤.
colors = ("r","g","b")

# ê° ì±„ë„ì„ ë°˜ë³µí•˜ë©´ì„œ ížˆìŠ¤í† ê·¸ëž¨ì„ ê³„ì‚°í•˜ê³  ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€í•©ë‹ˆë‹¤.
for i, channel in enumerate(colors):
    histogram = cv2.calcHist([image_rgb], # ì´ë¯¸ì§€
                        [i], # ì±„ë„ ì¸ë±ìŠ¤
                        None, # ë§ˆìŠ¤í¬ ì—†ìŒ
                        [256], # ížˆìŠ¤í† ê·¸ëž¨ í¬ê¸°
                        [0,256]) # ë²”ìœ„
    features.extend(histogram)

# ìƒ˜í”Œì˜ íŠ¹ì„± ê°’ìœ¼ë¡œ ë²¡í„°ë¥¼ ë§Œë“­ë‹ˆë‹¤.
observation = np.array(features).flatten()

# ì²˜ìŒ ë‹¤ì„¯ ê°œì˜ íŠ¹ì„±ì„ ì¶œë ¥í•©ë‹ˆë‹¤.
observation[0:5]
```
