```python
''' sum_3           교통지수
25% -1.229999	-1.032403e+00
50%	0.003758	-3.960821e-01
75%	1.167670	6.198185e-01
max	9.008577	1.452800e+01'''
# 열별로 사분위수 주기
def quant_sum(x):
    if x>= 9.008577:
        return 1
    elif x >= 1.167670  :
        return 2
    elif x >= 0.003758 :
        return 3
    else:
        return 4
        
def quant_trf(x):
    if x>= 1.452800e+01 :
        return 1
    elif x >= 6.198185e-01  :
        return 2
    elif x >= -3.960821e-01:
        return 3
    else:
        return 4

dfinter['tr_index']=dfinter['교통지수'].apply(quant_trf)
dfinter['sum_index']=dfinter['sum_3'].apply(quant_sum)
```
