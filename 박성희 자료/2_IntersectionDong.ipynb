```python
import pandas as pd
#dfsuit, zmytraffic 합치기
dfsuit=pd.merge(dfsuit,zmytraffic, how='inner',on='동')
```

```python
# 열뽑아서 series 만들기
s_sum=dfsuit['sum_3']
print(s_sum)
s_tr=dfsuit['교통지수']
print(s_tr)
```

```python
#교차분석 진행 하자.
#순위
# dfsuit.sort_values(by='sum_3',)
#x=sum_3 :-1.238667e-02  y='교통지수' :-0.392293


#apply 적용할 함수 정의 
def inter(x,y):
    if x>= -0.01238667 & y <= -0.392293 :
        return 4
    elif x< -0.01238667 & y <= -0.392293 :
        return 3
    elif x>= -0.01238667 & y > -0.392293 :
        return 1
    else :
        return 2
```

```python
def med_sum(x):
    if x>= -1.238667e-02:
        return 1
    else:
        return 0
def med_trf(x):
    if x<=-0.392293 :
        return 1
    else:
        return 0
```

```python
dfsuit['tr_index']=dfsuit['교통지수'].apply(med_trf)
dfsuit['sum_index']=dfsuit['sum_3'].apply(med_sum)
```

```python
def inter_g(s,m):
    if s==1 & m==1:
        return 1
```

```python
#교차분석 done : pandas , 두가지 열 apply
dfsuit['interg']=dfsuit.apply(lambda x:inter_g(x.tr_index,x.sum_index),axis=1)
```

```python
display(dfsuit)
```

```python
#그룹연산
grouped=dfsuit.groupby(['interg'])
print(grouped)
#그룹 분할
for key,group in grouped:
    print('1 dong',key)
    print('1 filter', group)

print('\n'*4)
#특정 그룹 골라서 추출
group1=grouped.get_group(1)
print(group1.tail())

display(group1)
```
